```python
import numpy as np
import pandas as pd

dataset =  pd.read_csv('Position_Salaries.csv')
display(dataset.head())

# Diferencia entre matriz y  vector
X = dataset.iloc[:, 1:2].values
y = dataset.iloc[:, 2].values

y
```

```python

# from sklearn.model_selection import train_test_split
# X_train, X_test, y_train, y_test = train_test_split(X, y, test_size = 0.2, random_state = 0)
```

### No podemos quitar elementos para predecir ya que no hay suficiente informacion y solo hay un dato por cada tipo

```python
import matplotlib.pyplot as plt
# Ajustar la regresion lineal con el dataset

from sklearn.linear_model import LinearRegression
lin_reg = LinearRegression()
lin_reg.fit(X, y)

#plt.scatter(X, y, color = 'red')


# Ajustar la regresion polinomica con el dataset
from sklearn.preprocessing import PolynomialFeatures

# transformar la matriz de caracteristicas X a X polineal, a√±adiendo X a la 2
poly_reg = PolynomialFeatures(degree = 2)

X_poly = poly_reg.fit_transform(X)
lin_reg_2 = LinearRegression()
lin_reg_2.fit(X_poly, y)


# VIsualizacion de los resultados del modelo lineal
plt.scatter(X,y,color='red')
plt.plot(X,lin_reg.predict(X), color='blue')
plt.xlabel('Posicion del empleado')
plt.ylabel('Sueldo en dolares')
plt.show()

```

### El modelo lineal de prediccion no es precisamente adecuado por cobrar mucho menos y algunos en valores negativos

```python

# VIsualizacion de los resultados del modelo lineal
plt.scatter(X,y,color='red')
plt.plot(X,lin_reg_2.predict(X_poly), color='blue')
plt.xlabel('Posicion del empleado')
plt.ylabel('Sueldo en dolares')
plt.show()
```

```python
# transformar la matriz de caracteristicas X a X polineal, a√±adiendo X a la 2
poly_reg_degree_4 = PolynomialFeatures(degree = 4)

X_poly3 = poly_reg_degree_4.fit_transform(X)
lin_reg_4 = LinearRegression()
lin_reg_4.fit(X_poly3, y)

# VIsualizacion de los resultados del modelo lineal
plt.scatter(X,y,color='red')
plt.plot(X,lin_reg_4.predict(X_poly3), color='blue')
plt.xlabel('Posicion del empleado')
plt.ylabel('Sueldo en dolares')
plt.show()
```

```python
# Indicaremos una secuencia de valores, en vez de valores discretos para que tome valores intermedios
X_grid = np.arange(min(X), max(X), 0.1)
X_grid = X_grid.reshape(len(X_grid),1) # vector de 90 filas y 1 columna

plt.scatter(X,y,color='red')
plt.plot(X_grid,lin_reg_4.predict(poly_reg_degree_4.fit_transform(X_grid)), color='blue')
plt.xlabel('Posicion del empleado')
plt.ylabel('Sueldo en dolares')
plt.show()
```
