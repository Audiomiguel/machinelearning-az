```python
# Regresion Linear Multiple

import numpy as np 
import matplotlib.pyplot as plt
import pandas as pd 

dataset = pd.read_csv('50_Startups.csv')

display(dataset.head())


X = dataset.iloc[:, :-1].values
y =dataset.iloc[:,4].values

from sklearn.model_selection import train_test_split

X_train, X_test, y_train, y_test = train_test_split(X, y, test_size=0.2, random_state=0)

# State debe ser una variable categorica
from sklearn.preprocessing import OneHotEncoder
from sklearn.compose import ColumnTransformer

# Seleccionamos donde esta la variable categorica
ct = ColumnTransformer(transformers=[('encoder', OneHotEncoder(), [3])], remainder='passthrough')

# Transforma la columna que yo indique a dummy
X = np.array(ct.fit_transform(X))

display(X)

```

```python
# Considerar que se movio el orden de las columnas

# Evitar la trampa de las variables dummy, podemos eliminar una columna, por ejemplo la primera

X= X[:, 1:]
print(X)
```

```python
from sklearn.model_selection import train_test_split
X_train, X_test, y_train, y_test = train_test_split(X, y, test_size = 0.2, random_state = 0)
```

```python
# Prediccion de los resultados en el conjunto de testing

from sklearn.linear_model import LinearRegression

regression = LinearRegression()
regression.fit(X_train, y_train)

y_pred = regression.predict(X_test)


# Consturir el model optimo de RLM utilizando la Eliminacion hacia atras
import statsmodels.formula.api as sm
tupla = (50,1)

print(X)

X = np.append( arr =  np.ones(tupla).astype(int), axis = 1, values = X) # 1 columna

print(X)

```

```python

# Parte con todas las columnas y luego elimina la menos significativa
X_opt =  X[:,[0,1,2,3,4,5]]
X_opt = np.array(X_opt, dtype = float)


# seleccionamos nivel de signficicaion
SL = 0.05 
```

## PASO 2 , Crear un modelo con todas las posibles variables predictoras

```python
# Minimos cuadrados ordinarios
import statsmodels.api as sm

regression_OLS = sm.OLS(endog = y , exog = X_opt).fit()
```

## Considera la variable predictora con el p-valor mas grande. Si P > SL, entonces vamos al paso de eliminar la variable predictora , sino fin

```python
display(regression_OLS.summary())
```

### Eliminamos la variable con el P mas elevado en este caso seria x1 y x2 (NY). Sin embargo X1 es una columna dummy, entonces no podemos tomarla

```python

# Parte con todas las columnas y luego elimina la menos significativa
X_opt =  X[:,[0,1,3,4,5]]
X_opt = np.array(X_opt, dtype = float)

regression_OLS = sm.OLS(endog = y , exog = X_opt).fit()
regression_OLS.summary()
```

```python
# X1 es la variable dummy de florida

# Parte con todas las columnas y luego elimina la menos significativa
X_opt =  X[:,[0,3,4,5]]
X_opt = np.array(X_opt, dtype = float)

regression_OLS = sm.OLS(endog = y , exog = X_opt).fit()
regression_OLS.summary()
```

```python
# Me dice que la variable x2 es la que tiene el P valor mucho mas elevado
X_opt =  X[:,[0,3,5]]
X_opt = np.array(X_opt, dtype = float)

regression_OLS = sm.OLS(endog = y , exog = X_opt).fit()
regression_OLS.summary()
```

### Si observamos, x2 supera a 0.05 , lo cual deberiamos eliminarlo. Sin embargo si hacemos ello, estariamos solo con una variable independiente y seria una regresion lineal simple 😆. Entonces lo haremos ya que este es el metodo de la eliminacion hacia atras

```python
# Me dice que la variable x2 es la que tiene el P valor mucho mas elevado
X_opt =  X[:,[0,3]]
X_opt = np.array(X_opt, dtype = float)

regression_OLS = sm.OLS(endog = y , exog = X_opt).fit()
regression_OLS.summary()
```
